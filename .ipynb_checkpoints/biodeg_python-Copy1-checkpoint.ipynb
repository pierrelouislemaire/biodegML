```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
#load the dataset
data = pd.read_csv("biodeg.csv", sep = " ")

# check shape of dataset
print("shape of the data:", data.shape)
data.head()
```

```python
# check data types of all columns
data.dtypes
```

```python
data["F03[C-O]"]=pd.Categorical(data["F03[C-O]"],ordered=False)
data["B01[C-Br]"]=pd.Categorical(data["B01[C-Br]"],ordered=False)
data["B03[C-Cl]"]=pd.Categorical(data["B03[C-Cl]"],ordered=False)
data["Y"]=pd.Categorical(data["Y"],ordered=False)
```

```python
#check missing data
data.isnull().sum()
```

```python
# target class count
print(data["Y"].value_counts())

sns.set(style="darkgrid")
per = data["Y"].value_counts()/len(data)*100
ax = sns.countplot(x="Y", data=data)
for c in ax.containers:  
    ax.bar_label(c, labels=[per[0], per[1]], label_type='edge')
plt.title("target variable")
```

```python
#Visualisation des variables qualitatives

fig, axs = plt.subplots(1,3, figsize=(15,3))
sns.histplot(x = "F03[C-O]", data = data, ax = axs[0])
sns.countplot(x = "B01[C-Br]", data = data, ax = axs[1])
sns.countplot(x = "B03[C-Cl]", data = data, ax = axs[2])
axs[0].title.set_text("Histogramme de F03[C-O]")
axs[1].title.set_text("Histogramme de B01[C-Br]")
axs[2].title.set_text("Histogramme de B03[C-Cl]")
```

```python
#Visualisation des variables quantitatives
fig, axs = plt.subplots(5,4, figsize=(15,10))
axs = axs.ravel()

colnames=[]
for c in data.columns:
    if data[c].dtype=="float64":
        colnames.append(c)

for col, ax in zip(colnames, axs):
    sns.histplot(data[col], kde=True, bins=50, ax=ax)
    ax.set_title("Histogramme de " + col)

plt.tight_layout()
```

```python
#boxplot
fig = plt.figure(figsize=(15,5))

plt.boxplot(data[colnames], labels = colnames)
plt.title("Boxplot des variables quantitatives")
plt.show()
```

```python
#normalisation des data en scaled_data
d = preprocessing.scale(data[colnames], axis=0)
scaled_data = pd.DataFrame(d, columns=data[colnames].columns)
print(scaled_data)
```

```python
#boxplot de scaled data
fig = plt.figure(figsize=(15,5))

plt.boxplot(scaled_data[colnames], labels = colnames)
plt.title("Boxplot des variables quantitatives")
plt.show()
```

```python
# heatmap des variables quantitatives
from pandas.plotting import scatter_matrix
corr= scaled_data.corr()

corr = corr[abs(corr) >= 0.7]
sns.heatmap(corr, annot=False)
plt.show()
```

Heatmap qui montre les corrÃ©lation forte entre les variables 

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
pca = PCA()
## Estimation, calcul des composantes principales
C = pca.fit(scaled_data).transform(scaled_data)
## DÃ©croissance de la variance expliquÃ©e
plt.plot(pca.explained_variance_ratio_)
plt.show()
```

```python
## distribution des composantes principales
plt.boxplot(C[:,0:20])
plt.show()
```

```python
## ReprÃ©sentation des individus en fonction de Y =0 ou 1
plt.figure(figsize=(5,5))
for i, j, nom in zip(C[:,0], C[:,1], data["Y"]):
    color = "red" if nom  else "blue"
    plt.plot(i, j, "o",color=color)
plt.axis((-8,8,-5,10))  
plt.show()
```

```python
## coordonnÃ©es et reprÃ©sentation des variables
coord1=pca.components_[0]*np.sqrt(pca.explained_variance_[0])
coord2=pca.components_[1]*np.sqrt(pca.explained_variance_[1])
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(1, 1, 1)
for i, j, nom in zip(coord1,coord2, scaled_data.columns):
    plt.text(i, j, nom)
    plt.arrow(0,0,i,j,color='black')
plt.axis((-1.2,1.2,-1.2,1.2))
# cercle
c=plt.Circle((0,0), radius=1, color='gray', fill=False)
ax.add_patch(c)
plt.show()
```

```python
from sklearn.cluster  import  KMeans
from  sklearn.metrics  import confusion_matrix
clust=KMeans(n_clusters=6)
clust.fit(scaled_data)
classe=clust.labels_
print(classe)
```

```python
## RepÃ©sentation des individus dans les coordonnÃ©es de l'acp.
plt.figure(figsize=(10,8))
plt.scatter(C[:,0], C[:,1], c=classe) 
plt.show()
```
